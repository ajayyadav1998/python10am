# break statement

```python
c = 5
while c > 0:
    if c == 2:
        print('byebye')
        break
    print('hello world',c)
    c -= 1
else:
    print('shall i run or not')
```

```python
c = 1
print('loop is starting from here')
while c <= 50:
    if c >= 10:
        break
    print('whola.... i am inside the loop')
    print('hello {}'.format(c))
    c += 1
else:
    print('bhot ho gya')
print('loop will end here')
```

# continue statement

```python
c = 5
while c > 0:
    if c == 1:
        print('chal rha hu')
        continue
    print('python ',c)
    c -=1
else:
    print('i will run')
```

```python
c = 5
while c >= 0:
    c -= 1
    if c == 1:
        print('chal rha hu')
        continue
    print('python ',c)
    c -=1
else:
    print('i will run')
```

```python
c = 1
while c <= 10:
    c += 1
    if c % 2:
        continue
    print(c)
else:
    print('chalo bye')
```

```python
row = int(input('row : '))
col = int(input('col : '))
matrix = [[0 for c in range(col)] for r in range(row)]

for r in range(row):
    for c in range(col):
        matrix[r][c] = int(input(f'matrix[{r}][{c}] : '))
                                                
                                                
for var in matrix:
    print('\t\t\t\t',*var)
```
